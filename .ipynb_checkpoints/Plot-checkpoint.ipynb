```python
%matplotlib inline
import pandas as pd
from IPython.core.display import HTML
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.DataFrame.from_csv('data1.csv', index_col=None)
```

```python
car = pd.DataFrame.from_csv('CAR around event.csv', index_col=None)
```

```python
df = df.replace([np.inf, -np.inf], 0)
```

```python
df = df.fillna(0)
```

```python
df[['News_ST', 'News_CE']] = df[['News_ST', 'News_CE']].astype(float)
```

```python
fig1 = df[['lgap','at_log']]
```

```python
fig1 = fig1.rename(columns={'lgap': 'Settle Period', 'at_log': 'Firm Size'})
```

```python
sns.set(style="white", color_codes=True)
```

```python
f1 = sns.JointGrid(x="Settle Period", y="Firm Size", data=fig1)
f1 = f1.plot(sns.regplot, sns.distplot)
f1.savefig("fig1.pdf", format='pdf', dpi=800)
```

```python
fig2 = car[['evttime','car']]
```

```python
fig2 = fig2.rename(columns={'evttime': 'Days Relative to Settlement Date 0', 'car': 'Cummulative Abnormal Return'})
```

```python
sns.set_style("whitegrid")
f2 = sns.barplot(x = "Days Relative to Settlement Date 0", y="Cummulative Abnormal Return", data=fig2, palette="Blues_d")
f2.figure.savefig("fig2.pdf", format='pdf', dpi=800)
```
